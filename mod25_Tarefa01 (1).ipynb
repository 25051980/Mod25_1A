1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. Método Mínimos Quadrados
    2. Gradiente descendente
    3. Newton Raphson
    4. CART (Árvore de decisão)

### Uso de Derivadas em Métodos Populares de Ciência de Dados

1. **Método dos Mínimos Quadrados**:
   - **Baseado em derivadas**: Este método utiliza a derivada da função de erro (soma dos quadrados dos erros) para encontrar os parâmetros que minimizam o erro. É amplamente usado em regressão linear.

2. **Gradiente Descendente**:
   - **Baseado em derivadas**: Este algoritmo utiliza o gradiente (derivadas parciais) para ajustar os parâmetros de um modelo, movendo-se na direção oposta ao gradiente para minimizar uma função de custo.

3. **Newton-Raphson**:
   - **Baseado em derivadas**: Este método utiliza tanto a primeira quanto a segunda derivada (hessiana) de uma função para encontrar zeros ou mínimos de forma mais eficiente do que o gradiente descendente simples.

4. **CART (Árvore de Decisão)**:
   - **Não utiliza derivadas**: Este método divide os dados em nós com base em medidas de impureza (como Gini ou entropia) e não depende de derivadas para tomar decisões.

Em resumo, os métodos baseados no uso de derivadas são o **Método dos Mínimos Quadrados**, **Gradiente Descendente** e **Newton-Raphson**, enquanto o **CART** não utiliza derivadas.


2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

O método dos **mínimos quadrados** é projetado para encontrar os parâmetros que minimizam o **Erro Quadrático Médio (EQM)**, ajustando o modelo de forma que a soma dos quadrados dos resíduos seja a menor possível. Isso ocorre porque ele resolve diretamente a equação que minimiza o EQM, garantindo que o valor obtido seja um **ótimo global** dentro do espaço de parâmetros do modelo. Qualquer outra solução resultaria em um EQM maior ou igual, mas nunca menor. Em modelos de regressão linear, os mínimos quadrados fornecem uma solução exata, desde que os pressupostos do modelo, como linearidade, homocedasticidade e ausência de multicolinearidade severa, sejam atendidos. A única forma de reduzir o EQM seria mudar a estrutura do modelo ou usar outro conjunto de dados, já que o método dos mínimos quadrados já atinge o menor EQM possível dado o modelo atual.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### Explicação do Código

Este código implementa o **método dos mínimos quadrados** para ajustar um modelo de regressão linear em um conjunto de dados simulados. Aqui está o que cada etapa faz:

1. **Geração dos dados simulados**: 
   - Um conjunto de 100 pontos é gerado para a variável explicativa (\(X\)) e a variável resposta (\(y\)) com um pouco de ruído aleatório para simular variações reais.

2. **Ajuste do modelo**:
   - O modelo de regressão linear é ajustado usando o `LinearRegression` da biblioteca `sklearn`. Ele calcula os coeficientes \(a\) (slope) e \(b\) (intercepto) que minimizam o **Erro Quadrático Médio (EQM)**.

3. **Cálculo do EQM**:
   - O EQM é calculado usando a função `mean_squared_error`, que mede a soma dos quadrados das diferenças entre os valores observados (\(y\)) e previstos (\(y_{pred}\)).

4. **Visualização**:
   - Os dados reais (\(X, y\)) são exibidos como pontos no gráfico.
   - A linha vermelha mostra o ajuste feito pelo modelo, representando a melhor aproximação linear dos dados.

5. **Resultados**:
   - O código imprime os coeficientes ajustados (slope e intercepto) e o valor do EQM, que indica o erro médio do modelo.

Este exemplo demonstra como o método dos mínimos quadrados encontra os parâmetros ótimos para minimizar o EQM e como ele ajusta uma linha reta aos dados observados.


```python

# Gerando dados simulados
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Variável explicativa
y = 3 * X.squeeze() + 7 + np.random.randn(100) * 2  # Resposta com ruído

# Ajustando o modelo de regressão linear (método dos mínimos quadrados)
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Calculando o EQM
mse = mean_squared_error(y, y_pred)

# Exibindo os resultados
print("Coeficiente angular (slope):", model.coef_[0])
print("Intercepto:", model.intercept_)
print("Erro Quadrático Médio (EQM):", mse)

# Visualizando os dados e o ajuste do modelo
plt.scatter(X, y, label="Dados reais", alpha=0.7)
plt.plot(X, y_pred, color="red", label="Ajuste do modelo (mínimos quadrados)")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Ajuste do Modelo de Regressão Linear")
plt.legend()
plt.show()
```

### Explicação dos Resultados

Os resultados exibidos representam o ajuste de um modelo de regressão linear pelo método dos mínimos quadrados:

1. **Coeficiente angular (slope)**:
   - O valor `2.91` indica que, para cada aumento unitário na variável explicativa \(X\), espera-se um aumento médio de aproximadamente 2.91 na variável resposta \(y\).

2. **Intercepto**:
   - O valor `7.43` representa o ponto onde a linha de regressão intercepta o eixo \(y\), ou seja, o valor previsto para \(y\) quando \(X = 0\).

3. **Erro Quadrático Médio (EQM)**:
   - O EQM de `3.226` mede a média das diferenças quadradas entre os valores reais (\(y\)) e os valores previstos (\(\hat{y}\)). Este valor reflete o quão bem o modelo ajusta os dados. Um EQM mais baixo indica um ajuste melhor.

4. **Gráfico**:
   - Os pontos azuis representam os dados reais.
   - A linha vermelha é o modelo ajustado, que minimiza o EQM. Ela demonstra a melhor aproximação linear dos dados fornecidos.

Esses resultados confirmam que o método dos mínimos quadrados foi bem-sucedido em ajustar um modelo linear aos dados observados.
